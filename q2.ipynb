```python
import nltk
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from nltk.probability import ConditionalFreqDist
%matplotlib inline
```

```python
file_en = open("data/text.en", "r")
file_jp = open("data/text.fr", "r")
#text_en = file_en.read()
#text_jp = file_jp.read()
list_en = [line[0:-1].split(' ') for line in file_en.readlines()]
list_jp = [line[0:-1].split(' ') for line in file_jp.readlines()]
```

```python
print(len(list_en))
print(list_en)
```

```python
print(len(list_jp))
print(list_jp)
```

```python
len_en = []
len_jp = []
for i in range(0, len(list_en)):
    len_en.append(len(list_en[i]))
    len_jp.append(len(list_jp[i]))
#print(len_en)
#print(len_jp)
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
plt.scatter(len_en, len_jp, alpha = 0.5)
```

```python
print(sum(len_en), sum(len_jp))
```

```python
dict_en = {}
for sen in list_en:
    for word in sen:
        dict_en[word] = 0
for sen in list_en:
    for word in sen:
        dict_en[word] += 1
print(len(dict_en))
print(len({k: v for k, v in dict_en.items() if v == 1}))
```

```python
dict_jp = {}
for sen in list_jp:
    for word in sen:
        dict_jp[word] = 0
for sen in list_jp:
    for word in sen:
        dict_jp[word] += 1
print(len(dict_jp))
print(len({k: v for k, v in dict_jp.items() if v == 1}))
```
